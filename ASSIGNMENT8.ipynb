```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Data
study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])

# Scatter Plots
plt.figure(figsize=(15, 5))

# Study Hours vs Exam Scores
plt.subplot(1, 3, 1)
plt.scatter(study_hours, exam_scores, color='b')
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')

# Sleep Hours vs Exam Scores
plt.subplot(1, 3, 2)
plt.scatter(sleep_hours, exam_scores, color='r')
plt.title('Sleep Hours vs Exam Scores')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Scores')

# Study Hours vs Sleep Hours
plt.subplot(1, 3, 3)
plt.scatter(study_hours, sleep_hours, color='g')
plt.title('Study Hours vs Sleep Hours')
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')

plt.tight_layout()
plt.show()

# Pearson correlation coefficients
study_exam_corr, _ = pearsonr(study_hours, exam_scores)
sleep_exam_corr, _ = pearsonr(sleep_hours, exam_scores)
study_sleep_corr, _ = pearsonr(study_hours, sleep_hours)

print("Pearson correlation coefficient between Study Hours and Exam Scores:", study_exam_corr)
print("Pearson correlation coefficient between Sleep Hours and Exam Scores:", sleep_exam_corr)
print("Pearson correlation coefficient between Study Hours and Sleep Hours:", study_sleep_corr)
```
