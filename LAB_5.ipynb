```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
N = 1000
vehicles_per_household = np.random.binomial(5, 0.4, N)
pmf = np.bincount(vehicles_per_household) / N
cdf = np.cumsum(pmf)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(range(6), pmf)
plt.xlabel('Number of Vehicles')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF)')

plt.subplot(1, 2, 2)
plt.plot(range(6), cdf, marker='o')
plt.xlabel('Number of Vehicles')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')

plt.tight_layout()
plt.show()
def probability(num_vehicles):
    return pmf[num_vehicles]
print("Probability of randomly selecting a household with 2 vehicles:", probability(2))
```

```python
import numpy as np
from scipy.stats import binom
N = 100
p = 0.05
k = 10
prob = binom.pmf(k, N, p)
print(f"P(X = {k}) ≈ {prob:.4f}")
prob = 1 - binom.cdf(14, N, p)
print(f"P(X ≥ {15}) ≈ {prob:.4f}")
expected = N * p
print(f"E(X) = {expected:.2f}")
```
