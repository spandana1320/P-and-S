```python
#QUESTION1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean = 50
std_dev = 5
num_samples = 1000
# Generate random data points from a normal distribution
data = np.random.normal(mean, std_dev, num_samples)
# Plot the histogram of the generated data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label="Data Histogram")
# Plot the PDF of the normal distribution
xmin, xmax = plt.xlim()  # Get the limits of the x-axis
x = np.linspace(xmin, xmax, 100)  # Generate x values for the PDF
p = norm.pdf(x, mean, std_dev)  # Calculate the corresponding y values (PDF)
plt.plot(x, p, 'k', linewidth=2, label="Normal Distribution PDF")
plt.title('Histogram and PDF of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()


#QUESTION2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# Parameters for the normal distribution
mean_test = 14
std_dev_test = 2.5
num_cases = 1000
# I. Calculate how many students scored between 14 and 16.5 (14 + 2.5)
prob_between_14_and_16_5 = norm.cdf(14 + 2.5, mean_test, std_dev_test) - norm.cdf(14, mean_test, std_dev_test)
students_between_14_and_16_5 = prob_between_14_and_16_5 * num_cases
# II. Calculate how many students scored below 12
prob_below_12 = norm.cdf(12, mean_test, std_dev_test)
students_below_12 = prob_below_12 * num_cases
# III. Calculate how many students scored above 19
prob_above_19 = 1 - norm.cdf(19, mean_test, std_dev_test)
students_above_19 = prob_above_19 * num_cases
print(f"Students who scored between 14 and 16.5: {students_between_14_and_16_5:.2f}")
print(f"Students who scored below 12: {students_below_12:.2f}")
print(f"Students who scored above 19: {students_above_19:.2f}")
x = np.linspace(mean_test - 4 * std_dev_test, mean_test + 4 * std_dev_test, 1000)
pdf_values = norm.pdf(x, mean_test, std_dev_test)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_values, label="Normal Distribution PDF", color='b', linewidth=2)
plt.fill_between(x, 0, pdf_values, where=((x >= 14) & (x <= (14 + 2.5))), color='green', alpha=0.3, label="Between 14 and 16.5")
plt.fill_between(x, 0, pdf_values, where=(x <= 12), color='red', alpha=0.3, label="Below 12")
plt.fill_between(x, 0, pdf_values, where=(x >= 19), color='orange', alpha=0.3, label="Above 19")
plt.title('Normal Distribution and Calculated Probabilities')
plt.xlabel('Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
